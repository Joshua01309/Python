```python


#  Definir per√≠odo de tiempo
start_date = '2022-01-01'
end_date = '2024-12-01'

def monthly_close(accion):
    data = yf.download(accion, start=start_date, end=end_date)
    if not data.empty:
        return data['Close'].resample('M').last()
    else:
        return pd.Series(name=accion)

# Obtener cierres mensuales
acciones = {
    '^IRX': 'IRX',
    '^GSPC': 'S&P500',
    'NVDA': 'NVDA',
    'COIN': 'COIN',
    'GC=F': 'Oro',
    'CL=F': 'Petr√≥leo',
    'EURUSD=X': 'EURUSD',
    'BTC-USD': 'Bitcoin',
    'VGT': 'VGT',
    'BND': 'BND'
}

series = []
for ticker, nombre in acciones.items():
    serie = monthly_close(ticker)
    serie.name = nombre
    series.append(serie)

# Crear DataFrame
cierres_mensuales = pd.concat(series, axis=1).ffill().dropna()
activos = cierres_mensuales.columns.tolist()

# Calcular retornos y estad√≠sticas
retornos_mensuales = cierres_mensuales.pct_change().dropna() * 100
estadisticas = pd.DataFrame({
    'Retorno Promedio': retornos_mensuales.mean(),
    'Desviaci√≥n Est√°ndar': retornos_mensuales.std(),
    'Coef. Variaci√≥n (Riesgo/Retorno)': retornos_mensuales.std() / retornos_mensuales.mean().abs(),
    'Desempe√±o (Retorno/Riesgo)': retornos_mensuales.mean() / retornos_mensuales.std()
})

# Generar pesos aleatorios que sumen 1 (100%)
np.random.seed(1)  # Semailla para reproducibilidad
pesos = np.random.random(len(estadisticas))
pesos /= pesos.sum()  # Normalizar para que sumen 1

# Agregar columna de pesos al DataFrame
estadisticas['Pesos'] = pesos

# --- Modificar la funci√≥n de estilos para incluir los pesos ---
def color_metrics(val, metric_name):
    if metric_name == 'Retorno Promedio':
        is_max = val == estadisticas[metric_name].max()
        is_min = val == estadisticas[metric_name].min()
    elif metric_name == 'Desviaci√≥n Est√°ndar':
        is_max = val == estadisticas[metric_name].max()
        is_min = val == estadisticas[metric_name].min()
    elif metric_name == 'Coef. Variaci√≥n (Riesgo/Retorno)':
        is_max = val == estadisticas[metric_name].max()
        is_min = val == estadisticas[metric_name].min()
    elif metric_name == 'Desempe√±o (Retorno/Riesgo)':
        is_max = val == estadisticas[metric_name].max()
        is_min = val == estadisticas[metric_name].min()
    elif metric_name == 'Pesos':
        is_max = val == estadisticas[metric_name].max()
        is_min = val == estadisticas[metric_name].min()

    color = ''
    if metric_name in ['Retorno Promedio', 'Desempe√±o (Retorno/Riesgo)', 'Pesos']:
        color = 'background-color: #90EE90' if is_max else 'background-color: #FFCCCB' if is_min else ''
    else:  # Para m√©tricas donde menor es mejor
        color = 'background-color: #90EE90' if is_min else 'background-color: #FFCCCB' if is_max else ''
    return color

# --- Actualizar los subsets en la aplicaci√≥n de estilos ---
styled_table = estadisticas.style\
    .applymap(lambda x: color_metrics(x, 'Retorno Promedio'), subset=['Retorno Promedio'])\
    .applymap(lambda x: color_metrics(x, 'Desviaci√≥n Est√°ndar'), subset=['Desviaci√≥n Est√°ndar'])\
    .applymap(lambda x: color_metrics(x, 'Coef. Variaci√≥n (Riesgo/Retorno)'), subset=['Coef. Variaci√≥n (Riesgo/Retorno)'])\
    .applymap(lambda x: color_metrics(x, 'Desempe√±o (Retorno/Riesgo)'), subset=['Desempe√±o (Retorno/Riesgo)'])\
    .applymap(lambda x: color_metrics(x, 'Pesos'), subset=['Pesos'])

# --- Agregar formato a los pesos ---
styled_table = styled_table.format({
    'Pesos': '{:.2%}',
    'Retorno Promedio': '{:.2f}%',
    'Desviaci√≥n Est√°ndar': '{:.2f}%'
})

# --- C√ìDIGO DE MATRICES ---
# Calcular matrices de correlaci√≥n y covarianza
correlaciones = retornos_mensuales.corr()
covarianzas = retornos_mensuales.cov()

# Crear tablas estilizadas para las matrices
def estilo_matriz(val):
    color = '#FFAAAA' if val < -0.5 else '#FFD700' if val < 0 else '#90EE90' if val > 0.5 else '#FFFFE0'
    return f'background-color: {color}'

# Matriz de correlaci√≥n estilizada
corr_estilizada = correlaciones.style\
    .applymap(estilo_matriz)\
    .format('{:.2f}')\
    .set_caption('Matriz de Correlaci√≥n entre Activos')\
    .set_table_styles([{
        'selector': 'th',
        'props': [('font-size', '10pt'), ('background-color', '#404040'), ('color', 'white')]
    }])

# Matriz de covarianza estilizada
cov_estilizada = covarianzas.style\
    .background_gradient(cmap='RdYlGn', axis=None)\
    .format('{:.2f}')\
    .set_caption('Matriz de Covarianza entre Activos (en %¬≤)')\
    .set_table_styles([{
        'selector': 'th',
        'props': [('font-size', '10pt'), ('background-color', '#404040'), ('color', 'white')]
    }])

# --- SECCI√ìN DE VISUALIZACI√ìN ---
# Mostrar datos crudos
print("="*90)
print("Cierres mensuales alineados (primeras 5 filas):")
print(cierres_mensuales.head())

# Mostrar tabla estad√≠stica estilizada
from IPython.display import display
print("\n" + "="*90)
print("Estad√≠sticas Comparativas con Formato Condicional:")
print("Verde = Mejor en categor√≠a | Rojo = Peor en categor√≠a")
display(styled_table)

# Mostrar gr√°fico de pastel de pesos
print("\n" + "="*90)
print("üß©  Pesos en el Portafolio (pesos aleatorios)")
plt.figure(figsize=(10, 8))
explode = [0.1 if w == max(pesos) else 0 for w in pesos]
plt.pie(pesos,
        labels=estadisticas.index,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True,
        colors=plt.cm.Paired.colors)
plt.title('Composici√≥n del Portafolio Aleatorio', fontsize=14)
plt.axis('equal')
plt.show()

# Mostrar informaci√≥n del per√≠odo
print("\n" + "="*90)
print(f"üìÖ Per√≠odo analizado: {cierres_mensuales.index[0].strftime('%Y-%m-%d')} a {cierres_mensuales.index[-1].strftime('%Y-%m-%d')}")
print(f"üïí N√∫mero de meses analizados: {len(cierres_mensuales)}")

# Mostrar matrices de correlaci√≥n y covarianza
print("\n" + "="*90)
print("üìä Matriz de Correlaci√≥n:")
display(corr_estilizada)
print("\n" + "="*90)
print("üìà Matriz de Covarianza:")
display(cov_estilizada)

# --- SECCI√ìN PORTADAFOLIO 1 ---
print("\n" + "="*90)
print("üì¶ PORTAFOLIO 1 - M√©tricas Avanzadas")

# 1. Retorno del Portafolio
retorno_portafolio = (estadisticas['Pesos'] * estadisticas['Retorno Promedio']).sum()
print(f"\nüéØ Retorno Esperado: {retorno_portafolio:.2f}%")

# 2. Matriz de Markowitz (C√°lculo vectorizado)
matriz_markowitz = np.outer(estadisticas['Pesos'], estadisticas['Pesos']) * covarianzas
matriz_markowitz_display = pd.DataFrame(matriz_markowitz,
                                        index=estadisticas.index,
                                        columns=estadisticas.index)

print("\nüî∑ Matriz de Markowitz (Primeros 5 elementos):")
display(matriz_markowitz_display.iloc[:5, :5].style.format("{:.4f}"))

# 3. Riesgo del Portafolio (Desviaci√≥n est√°ndar)
riesgo_portafolio = np.sqrt(matriz_markowitz.sum().sum())
print(f"\n‚ö†Ô∏è Riesgo del Portafolio: {riesgo_portafolio:.2f}%")

# 4. Desempe√±o del Portafolio (Ratio de Sharpe)
desempeno_portafolio = retorno_portafolio / riesgo_portafolio
print(f"\nüöÄ Desempe√±o (Ratio de Sharpe): {desempeno_portafolio:.2f}")

# 5. Desglose de Contribuciones
print("\nüìå Contribuci√≥n por Activo:")
contribuciones = pd.DataFrame({
    'Activo': estadisticas.index,
    'Peso': estadisticas['Pesos'],
    'Retorno Individual': estadisticas['Retorno Promedio'],
    'Contribuci√≥n Retorno': estadisticas['Pesos'] * estadisticas['Retorno Promedio'],
    'Contribuci√≥n Riesgo': np.diag(matriz_markowitz)
})
display(contribuciones.style.format({
    'Peso': '{:.2%}',
    'Retorno Individual': '{:.2f}%',
    'Contribuci√≥n Retorno': '{:.2f}%',
    'Contribuci√≥n Riesgo': '{:.4f}'
}))

print("\n" + "="*90)
# --- SECCI√ìN PORTADAFOLIO 2 ---

from pyomo.environ import *

retornos = retornos_mensuales.values.T  # Matriz de retornos: [activos √ó periodos]
n_activos = retornos.shape[0] # Numero de activos
n_periodos = retornos.shape[1] # Numero de meses
retorno_medio = retornos_mensuales.mean().values # Retorno esperado del activo
n_activos = len(activos)

# Creamos el modelo
model = ConcreteModel()

# Definimos las variables de decisi√≥n
model.w = Var(range(n_activos), within=NonNegativeReals)  # Pesos de cada activo
model.d = Var(range(n_periodos), within=NonNegativeReals) # Desviaciones absolutas

# Definimos la funci√≥n objetivo
model.obj = Objective(expr = (1/n_periodos)*sum(model.d[t] for t in range(n_periodos)), sense=minimize)

# Definimos las restricciones
model.con1 = Constraint(expr= sum(model.w[i] for i in range(n_activos)) == 1) # Restricci√≥n 1: Suma de pesos = 100%
model.con2 = ConstraintList()
for t in range(n_periodos):
    expr = sum(model.w[i]*(retornos[i,t] - retornos_mensuales.mean().values[i])
              for i in range(n_activos))
    model.con2.add(expr <= model.d[t])   # d[t] >= R_t - E[R]
    model.con2.add(-expr <= model.d[t])  # d[t] >= -(R_t - E[R])

# Creamos un solver, por ejemplo GLPK
solver = SolverFactory('glpk', executable='/usr/bin/glpsol')# SolverFactory('glpk')
results = solver.solve(model)

# Mostrar resultados
print("\n" + "="*60)
print("üîç Resultados Portafolio 2 - Minimizaci√≥n de Riesgo (metodo lineal MAD)")
print("‚ïê"*60)

# Mostrar pesos √≥ptimos con formato
print("\n‚öñÔ∏è Pesos √ìptimos (M√≠nimo Riesgo):")
for i, activo in enumerate(activos):
    print(f"‚Ä¢ {activo}: {model.w[i]():.2%}")

print("\n‚öñÔ∏è Pesos √ìptimos (M√≠nimo Riesgo):")
pesos_optimos = [model.w[i]() for i in range(n_activos)]  # Extraer pesos

# Crear gr√°fico de pastel
plt.figure(figsize=(10, 8))
explode = [0.1 if w == max(pesos_optimos) else 0 for w in pesos_optimos]
colors = plt.cm.Paired.colors  # Mismo esquema de colores que Portafolio 1

plt.pie(pesos_optimos,
        labels=activos,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True,
        colors=colors,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'})

plt.title('Composici√≥n del Portafolio de M√≠nimo Riesgo', fontsize=14, pad=20)
plt.axis('equal')

print("\nüìä Gr√°fico de Distribuci√≥n:")
plt.show()

print(f"\nüìâ Valor optimo de Z (min Riesgo): {model.obj():.2f}%")
print(f"üéØ Retorno Esperado: {sum(model.w[i]()*retornos_mensuales.mean().values[i] for i in range(n_activos)):.2f}%")


print("\n" + "="*90)
# --- SECCI√ìN PORTADAFOLIO 3 ---

from pyomo.environ import *

# Creamos el modelo
model_sharpe = ConcreteModel()

# Definimos las variables de decisi√≥n
# Variables transformadas (Charnes-Cooper)
model_sharpe.y = Var(range(n_activos), within=NonNegativeReals)  # y_i = w_i * t
model_sharpe.t = Var(within=NonNegativeReals)                    # t = 1/MAD
model_sharpe.d = Var(range(n_periodos), within=NonNegativeReals)

# Definimos la funci√≥n objetivo
model_sharpe.obj = Objective(expr = sum(model_sharpe.y[i] * retorno_medio[i] for i in range(n_activos)),  sense=maximize)

# Definimos las restricciones
# Restricciones de normalizaci√≥n
model_sharpe.con1 = Constraint(expr = sum(model_sharpe.y[i] for i in range(n_activos)) == model_sharpe.t)
model_sharpe.con2 = Constraint(expr = (1/n_periodos)*sum(model_sharpe.d[t] for t in range(n_periodos)) <= 1)
# Restricciones de desviaciones absolutas (MAD)
model_sharpe.con3 = ConstraintList()
for t in range(n_periodos):
    expr = sum(model_sharpe.y[i]*(retornos[i,t] - retorno_medio[i]) for i in range(n_activos))
    model_sharpe.con3.add(expr <= model_sharpe.d[t])
    model_sharpe.con3.add(-expr <= model_sharpe.d[t])

# Creamos un solver, por ejemplo GLPK
solver = SolverFactory('glpk', executable='/usr/bin/glpsol')# SolverFactory('glpk')
results = solver.solve(model_sharpe)

# Mostrar resultados
print("\n" + "="*60)
print("üîç Resultados Portafolio 3 - Maximizacion de Desempe√±o (metodo lineal MAD)")
print("‚ïê"*60)

# 8. Calcular pesos originales y m√©tricas
if model_sharpe.t() > 0:
    pesos_sharpe = [model_sharpe.y[i]() / model_sharpe.t() for i in range(n_activos)]
    mad_sharpe = 1 / model_sharpe.t()
    retorno_sharpe = sum(pesos_sharpe[i] * retorno_medio[i] for i in range(n_activos))
else:
    pesos_sharpe = [0]*n_activos
    mad_sharpe = 0
    retorno_sharpe = 0

# 9. Mostrar resultados
print("\n‚öñÔ∏è Pesos √ìptimos (M√°ximo Desempe√±o):")
for i, activo in enumerate(activos):
    print(f"‚Ä¢ {activo}: {pesos_sharpe[i]:.2%}")

print(f"\nüìä M√©tricas:")
print(f"üìà Valor optimo de Z (max desempe√±o): {retorno_sharpe:.2f}%")
print(f"üìâ (Riesgo): {mad_sharpe:.2f}%")
print(f"üöÄ Ratio de Sharpe: {retorno_sharpe/mad_sharpe if mad_sharpe !=0 else 0:.2f}")

# 10. Gr√°fico de pastel
plt.figure(figsize=(10,8))
explode = [0.1 if w == max(pesos_sharpe) else 0 for w in pesos_sharpe]
plt.pie(pesos_sharpe,
        labels=activos,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode,
        shadow=True,
        colors=plt.cm.Paired.colors)
plt.title('Portafolio 3 - M√°ximo Desempe√±o', fontsize=14)
plt.axis('equal')
plt.show()
```
