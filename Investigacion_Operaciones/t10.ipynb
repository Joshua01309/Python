```python
# Importamos las bibliotecas necesarias
from pyomo.environ import *

# Creamos el modelo
model = ConcreteModel()

# Definimos las variables de decisión
model.x_12 = Var(within=NonNegativeReals)
model.x_13 = Var(within=NonNegativeReals)
model.x_14 = Var(within=NonNegativeReals)
model.x_25= Var(within=NonNegativeReals)
model.x_23 = Var(within=NonNegativeReals)
model.x_32 = Var(within=NonNegativeReals)
model.x_35 = Var(within=NonNegativeReals)
model.x_36 = Var(within=NonNegativeReals)
model.x_34 = Var(within=NonNegativeReals)
model.x_46= Var(within=NonNegativeReals)
model.x_57 = Var(within=NonNegativeReals)
model.x_56 = Var(within=NonNegativeReals)
model.x_65 = Var(within=NonNegativeReals)
model.x_67 = Var(within=NonNegativeReals)
model.x_71 = Var(within=NonNegativeReals)




# Definimos la función objetivo
model.obj = Objective(expr=model.x_71  , sense=maximize)

# Definimos las restricciones
# Complete con la restricción 1:
model.con1 = Constraint(expr= model.x_12 + model.x_13 + model.x_14 - model.x_71 == 0)
model.con2 = Constraint(expr= model.x_25 - model.x_23 - model.x_12 - model.x_32  == 0)
model.con3 = Constraint(expr= model.x_35 + model.x_32 + model.x_36 + model.x_34 - model.x_23 - model.x_13 == 0)
model.con4 = Constraint(expr= model.x_46 - model.x_14 - model.x_34  == 0)
model.con5 = Constraint(expr= model.x_57 + model.x_56 - model.x_25 - model.x_36 - model.x_65 == 0)
model.con6 = Constraint(expr= model.x_67 + model.x_65 - model.x_56 - model.x_36 - model.x_46 == 0)
model.con7 = Constraint(expr= model.x_71 - model.x_57 - model.x_67  == 0)


model.con8 = Constraint(expr= model.x_12 <= 5)
model.con9 = Constraint(expr= model.x_23 <= 2)
model.con10 = Constraint(expr= model.x_13 <= 6)
model.con11 = Constraint(expr= model.x_14 <= 5)
model.con12 = Constraint(expr= model.x_25 <= 3)
model.con13 = Constraint(expr= model.x_32 <= 2)
model.con14 = Constraint(expr= model.x_34 <= 3)
model.con15 = Constraint(expr= model.x_35 <= 3)
model.con16 = Constraint(expr= model.x_36 <= 7)
model.con17 = Constraint(expr= model.x_46 <= 5)
model.con18 = Constraint(expr= model.x_56 <= 1)
model.con19 = Constraint(expr= model.x_57 <= 8)
model.con20 = Constraint(expr= model.x_65 <= 1)
model.con21 = Constraint(expr= model.x_67 <= 7)


# Creamos un solver, por ejemplo GLPK
solver = SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()# SolverFactory('glpk')
# # Resolvemos el problema
# solver.solve(model)

# Mostramos los resultados
print(f"Valor óptimo de x_12: {model.x_12()}")
print(f"Valor óptimo de x_13: {model.x_13()}")
print(f"Valor óptimo de x_14: {model.x_14()}")
print(f"Valor óptimo de x_71: {model.x_71()}")
print(f"Valor óptimo de x_25: {model.x_25()}")
print(f"Valor óptimo de x_23: {model.x_23()}")
print(f"Valor óptimo de x_32: {model.x_32()}")
print(f"Valor óptimo de x_35: {model.x_35()}")
print(f"Valor óptimo de x_36: {model.x_36()}")
print(f"Valor óptimo de x_34: {model.x_34()}")
print(f"Valor óptimo de x_46: {model.x_46()}")
print(f"Valor óptimo de x_57: {model.x_57()}")
print(f"Valor óptimo de x_56: {model.x_56()}")
print(f"Valor óptimo de x_65: {model.x_65()}")
print(f"Valor óptimo de x_67: {model.x_67()}")


print(f"Valor óptimo de Z: {model.obj()}")


model.display()
```
